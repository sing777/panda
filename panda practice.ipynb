 Differ concat and merge

1. concat accept list of df & merge accept only 2 df.
2. index reset not done in concat & by default index reset is done in merge.
3. concat having join ['outer','inner'] arg & merge having how ={'left', 'right', 'outer', 'inner', 'cross'} arg
4. concat will join df based axes [row, columns] & merge will join df based on data or elements.
5. concat having axis as arg & merge having how = {}

```python
# Write a Pandas program to extract years between 1800 to 2200 from the specified
# column of a given DataFrame.
df = pd.DataFrame({
'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
'year': ['year 1800','year 1700','year 2300', 'year 1900', 'year 2200']
})
df
```

```python
def extract(string):
 var=re.findall(r"18[\d]{2}|19[\d]{2}|20[\d]{2}|21[\d]{2}|22[\d]{2}",string)
 return var
df["new_year"]=df["year"].apply(extract)
df
```

```python
def extract(string):
 var=re.findall(r"[1800]{1}[1900]{1}[2200]{1}",string)
 return var
df['new_year']=df["year"].apply(extract)
df
```
